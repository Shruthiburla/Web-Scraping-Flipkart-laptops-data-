```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
processor=[]
ram=[]
os=[]
storage=[]
display=[]
warranty=[]
rating=[]
discount_price=[]
price=[]
title=[]
```

```python
for i in range(1,19):
    url='https://www.flipkart.com/search?q=iphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(i)
    
    page=requests.get(url)
    htmlcode=page.text
    
    soup=BeautifulSoup(htmlcode)
    
    rams=soup.find_all('li',attrs={'class' : 'rgWa7D'})    
    for item in rams:
        ram.append(item.text)  
    
    ratings=soup.find_all('div',attrs={'class' : '_3LWZlK'})    
    for item in ratings:
        rating.append(item.text)  
    
    discount_prices=soup.find_all('div',attrs={'class' : '_30jeq3 _1_WHN1'})    
    for item in discount_prices:
        discount_price.append(item.text) 
        
    processors=soup.find_all('div',attrs={'class' : '_3I9_wc _27UcVY'})    
    for item in processors:
        processor.append(item.text)   
    
    oss=soup.find_all('li',attrs={'class' : 'rgWa7D'})    
    for item in oss:
        os.append(item.text)  
        
    displays=soup.find_all('li',attrs={'class' : 'rgWa7D'})    
    for item in displays:
        display.append(item.text)   
        
    storages=soup.find_all('li',attrs={'class' : 'rgWa7D'})    
    for item in storages:
        storage.append(item.text)   
        
    warrantys=soup.find_all('li',attrs={'class' : 'rgWa7D'})    
    for item in warrantys:
        warranty.append(item.text)   
        
    ratings=soup.find_all('div',attrs={'class' : '_3LWZlK'})    
    for item in ratings:
        rating.append(item.text)   
        
    prices=soup.find_all('div',attrs={ 'class' : '_3I9_wc _27UcVY'})    
    for item in prices:
        price.append(item.text)  
        
    titles=soup.find_all('div',attrs={'class' : '_4rR01T'})    
    for item in titles:
        title.append(item.text)      
          
```

```python
print(len(title))
print(len(processor))
print(len(ram))
print(len(os))
print(len(storage))
print(len(display))
print(len(warranty))
print(len(price))
print(len(rating))
print(len(discount_price))
```

```python
d={
   'Title':title[0:422],
   'RAM':ram[0:422],
   'os':os[0:422],
   'Storage':storage[0:422],
   'Display':display[0:422],
   'Warranty':warranty[0:422], 
   'discount_price':discount_price[0:422],
   'ratings':rating[0:422]}
```
